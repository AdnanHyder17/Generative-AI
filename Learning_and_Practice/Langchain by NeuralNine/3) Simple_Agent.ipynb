```python
import requests
from dotenv import load_dotenv

load = load_dotenv()

from langchain.agents import create_agent # change langchain.chat_models with this to create agents 
from langchain.tools import tool
```

```python
@tool('get_weather', return_direct=False, description='Get the current weather for a given city')
def get_weather(city: str) -> str:
    response = requests.get('https://wttr.in/{City}?format=3'.format(City=city))
    return response.text
```

```python
agent = create_agent(
    model = "gpt-4o-mini",
    tools = [get_weather],
    system_prompt = "You're a helpful assistant that provides weather information."
)
```

```python
response = agent.invoke({
    'messages': [
        {
            'role': 'user',
            'content': 'What is the weather in New York City?'
        }
    ]
})
```
