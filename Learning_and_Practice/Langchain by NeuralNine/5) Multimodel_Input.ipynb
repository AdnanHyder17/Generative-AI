```python
from dotenv import load_dotenv
load_dotenv()

from langchain.chat_models import init_chat_model
from langchain.messages import HumanMessage
```

```python
model = init_chat_model('gpt-4', temperature=0)
```

```python
message_passing_img_url = {
    'role': 'user',
    'content': [
        {'type': 'text', 'text': 'Describe the image.'},
        {'type': 'image', 'url': 'https://example.com/image.jpg'}
    ]
}

# Alternative message with HumanMessage class
message_passing_img_url = HumanMessage(
    content= [
        {'type': 'text', 'text': 'Describe the image.'},
        {'type': 'image', 'url': 'https://example.com/image.jpg'}
    ]
)
```

```python
from base64 import b64encode

message_passing_img_locally = {
    'role': 'user',
    'content': [
        {'type': 'text', 'text': 'Describe the image.'},
        {
            'type': 'image', 
            'base64': b64encode(open('image.jpg', 'rb').read()).decode(),
            'mime_type': 'image/jpeg'
        }
    ]
}

# Alternative message with HumanMessage class
message_passing_img_locally = HumanMessage(
    content= [
        {'type': 'text', 'text': 'Describe the image.'},
        {
            'type': 'image', 
            'base64': b64encode(open('image.jpg', 'rb').read()).decode(),
            'mime_type': 'image/jpeg'
        }
    ]
)
```

```python
response = model.invoke([message_passing_img_url])
print(response['content'])

response = model.invoke([message_passing_img_locally])
print(response['content'])
```
